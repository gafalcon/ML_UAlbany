```python
import numpy as np
import matplotlib.pyplot as plt
```

# Read data from file

```python
data = []
with open('hw2.dat', 'r') as f:
    for l in f:
        data.append([ float(n) for n in l.split(' ') ])
```

```python
data = np.array(data)
x = data[:,0]
y = data[:,1]
```

# Create polynomial X matrix

```python
def create_poly_matrix(x, dimension):
    X = np.zeros( ((dimension+1), len(x)) )
    for i in range(0, dimension+1):
        X[i,:] = np.power(x,i)
    return X
```

# Calculate Regression

```python
def regression(X, y):
    cor = X @ X.T
    lpm = np.linalg.inv(cor) @ X
    w = lpm @ y
    return w
```

# Calculate Predictions and Error

```python
def predict(X, w, y):
    pred = X.T @ w
    err = y - pred
    total_err = err.sum()
    return pred, total_err
```

# Plot results

```python
def plot_regression(x, y, pred, d):
    plt.plot(x,y, 'ro', x, pred, 'g^')
    plt.title("{}-degree polynomial regression".format(d))
    plt.xlabel("x")
    plt.ylabel("y")
```

# Polynomial d = 1

```python
X = create_poly_matrix(x, 1)
w = regression(X, y)
pred, total_err = predict(X, w, y)
print("Total Error:",total_err)
plot_regression(x,y,pred,1)
```
