```python
import numpy as np
import matplotlib.pyplot as plt
```

# Read data from file

```python
data = []
with open('hw2.dat', 'r') as f:
    for l in f:
        data.append([ float(n) for n in l.split(' ') ])
```

```python
data = np.array(data)
x = data[:,0]
y = data[:,1]
```

# Create matrix

```python
def create_poly_matrix(x, dimension):
    """Creates polynomial matrix"""
    X = np.zeros( ((dimension+1), len(x)) )
    for i in range(0, dimension+1):
        X[i,:] = np.power(x,i)
    return X
```

```python
def regression(X, y):
    """Calculates w that best predict y"""
    cor = X @ X.T
    cor_inv = np.linalg.inv(cor)
    lpm = cor_inv @ X
    w = lpm @ y
    return w, cor,cor_inv
```

```python
def predict(X, w, y):
    pred = X.T @ w
    err = y - pred
    total_err = err.dot(err)
    #print("total error", total_err)
    return pred, err, total_err
```

```python
def plot_regression(x, y, pred, d):
    plt.plot(x,y, 'ro', x, pred, 'g^')
    plt.title("{}-degree polynomial regression".format(d))
    plt.xlabel("x")
    plt.ylabel("y")
```

# Calculate new inverse correlation matrix

```python
def update_correlation_matrix(X_new, cor_inv):
    """Calculates new cor matrix and inv cor matrix based on previous matrix"""
    cor_new = X_new @ X_new.T
    x_cor = cor_new[:-1, -1:]
    d = cor_new[-1,-1]
    a = cor_inv @ x_cor
    b = x_cor.T @ cor_inv
    den = (d - ((x_cor.T @ cor_inv) @ x_cor))
    cor_new_inv = np.zeros(cor_new.shape)
    cor_new_inv[-1, -1] = 1/den
    cor_new_inv[:-1, -1:] = (-a)/den
    cor_new_inv[-1:, :-1] = (-b)/den
    cor_new_inv[:-1, :-1] = ((a @ b)/den) + cor_inv
    return cor_new, cor_new_inv
```

# Calculate LOO

```python
def average_loo(X, y, cor_inv, w):
    """
    Calculate the loo for every col of X
    """
    acum = 0
    n = X.shape[1]
    for i in range(n):
        y_i = y[i]
        x_i = X[:,i]
        acum += ((y_i - (x_i @ w)) **2) / ((1 - (x_i @ cor_inv) @ x_i)**2)
    return acum/n
```

# Recursive LLSE

```python
X = create_poly_matrix(x, 1) #create 1-polynomial matrix
dimension = 1
w, cor, cor_inv = regression(X,y) #Regression with 1-polynomial
pred, err, total_err = predict(X, w, y)
plt.figure(figsize=(20, 40))
plt.subplot(9, 3, 1)
plot_regression(x, y, pred, 1)
avg_loo = average_loo(X, y, cor_inv, w)
loos = [avg_loo] #store loo error in list

for dimension in range(2, 28): #iterate to create polynomials with power 2 to 27
    plt.subplot(9, 3, dimension)
    x_new = np.power(X[1,:],dimension) #New power dimension
    
    # One way to calculate new w based on previous inverse cor and previous w
    #w0 = x_new.dot(err) / (x_new.dot(x_new) - ((((x_new @ X.T) @ cor_inv) @ X) @ x_new))
    #print("w0", w0)
    #w_hat = w - (w0 * ((cor_inv @ X) @ x_new))
    #w_new = np.append(w_hat, w0)
    
    #Calculate new w by calculating first the new inverse cor matrix based on previous inv cor
    X_new = np.concatenate((X, np.expand_dims(x_new, axis=0)), axis= 0) #add dimension to X matrix
    cor, cor_inv = update_correlation_matrix(X_new, cor_inv) #calculate new cor and inverse cor
    lpm = cor_inv @ X_new
    w_new = lpm @ y #new w weights
    
    #Calculate error and plot
    pred, err, total_err = predict(X_new, w_new, y)
    plot_regression(x, y, pred, dimension)
    
    X = X_new
    w = w_new

    #Calculate loo and add it to list
    avg_loo = average_loo(X, y, cor_inv, w)
    loos.append(avg_loo)
```

# Plot LOO Error for each polynomial

```python
plt.plot(list(range(1, 28)), loos)
plt.title("Avg LOO Error vs polynomial degree")
plt.xlabel("Polynomial Degree")
plt.ylabel("Avg LOO Error")
```

```python
i_min = np.argmin(loos)
print("Polynomial with least error: ", i_min + 1, " loo error: ", loos[i_min])
```
